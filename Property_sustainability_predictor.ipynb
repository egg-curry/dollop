Applying ML

```python
x=data.drop(columns=['rent'])
y=data['rent']
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

Applying Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output = False, handle_unknown='ignore'), ['locality','city']), 
                                       remainder = 'passthrough', force_int_remainder_cols=False)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression(positive=False)
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

Applying LASSO (L1 Regularisation)

```python
Lasso=Lasso(positive=False)
```

```python
pipe = make_pipeline(column_trans, scaler, Lasso)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
r2_score(y_test, y_pred_lasso)
```

Applying Ridge (L2 Regularisation)

```python
Ridge=Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, Ridge)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_ridge=pipe.predict(x_test)
r2_score(y_test, y_pred_ridge)
```
